```python
import pandas as pd
```

```python
movies = pd.read_csv('./dataset/movies25.csv')
ratings = pd.read_csv('./dataset/ratings25.csv')
links = pd.read_csv('./dataset/links25.csv')

movies.shape, links.shape
```

```python
# merge movies and links by movieId
movies_links = pd.merge(movies, links, on='movieId')
movies_links.shape
```

```python
movies_links.head()
```

```python
movies_links[['title', 'year']] = movies_links['title'].str.extract(r'^(.*) \((\d{4})\)$')
```

```python
movies_links
```

```python
movies_links.isnull().sum()
```

```python
# show rows where null values
movies_links[movies_links['year'].isnull()]
```

```python
# get the list of ids of the rows where year is null
null_ids = movies_links[movies_links['year'].isnull()].index
```

```python
ratings = ratings[~ratings['movieId'].isin(null_ids)]
movies_links = movies_links[~movies_links['year'].isnull()]
```

```python
movies_links.to_csv('./dataset/movies_complete.csv', index=False)
ratings.to_csv('./dataset/ratings_complete.csv', index=False)
```

```python
# from ratings, get the list of the unique userId and save it in a dataframe
users = pd.DataFrame(ratings['userId'].unique(), columns=['userId'])
users.head()
```

```python
users.to_csv('./dataset/users.csv', index=False)
```
