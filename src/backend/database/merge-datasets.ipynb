```python
import pandas as pd
```

```python
movies = pd.read_csv('data/movies25.csv')
ratings = pd.read_csv('data/ratings25.csv')
links = pd.read_csv('data/links25.csv')

movies.shape, links.shape
```

```python
# merge movies and links by movieId
movies_links = pd.merge(movies, links, on='movieId')
movies_links.shape
```

```python
movies_links.head()
```

```python
movies_links[['title', 'year']] = movies_links['title'].str.extract(r'^(.*) \((\d{4})\)$')
```

```python
movies_links
```

```python
movies_links.isnull().sum()
```

```python
# show rows where null values
movies_links[movies_links['year'].isnull()]
```

```python
# get the list of ids of the rows where year is null
null_ids = movies_links[movies_links['year'].isnull()].index
```

```python
ratings = ratings[~ratings['movieId'].isin(null_ids)]
movies_links = movies_links[~movies_links['year'].isnull()]
```

```python
movies_links.to_csv('data/movies_complete.csv', index=False)
ratings.to_csv('data/ratings_complete.csv', index=False)
```

```python
# from ratings, get the list of the unique userId and save it in a dataframe
users = pd.DataFrame(ratings['userId'].unique(), columns=['userId'])
users.head()
```

```python
users.to_csv('data/users.csv', index=False)
```

```python
movies = pd.read_csv('data/movies_complete.csv')
ratings = pd.read_csv('data/ratings25.csv')
metadata = pd.read_csv('data/movies_metadata.csv')

movies.shape, ratings.shape, metadata.shape
```

```python
# in movies, tmdbId convert to int
movies['tmdbId'] = movies['tmdbId'].fillna(0).astype(int)
# remove rows where tmdbId is 0
movies = movies[movies['tmdbId'] != 0]
movies.head()
```

```python
metadata.head()
```

```python
metadata = metadata[['id', 'overview', 'vote_average', 'poster_path']]
```

```python
metadata.isnull().sum()
```

```python
metadata.dropna(inplace=True)
```

```python
# metadata id as int
metadata['id'] = metadata['id'].astype(int)
```

```python
# keep rows in movies where tmdbId is in metadata id and add overview and vote_average
movies = pd.merge(movies, metadata, left_on='tmdbId', right_on='id')
movies.shape
```

```python
movies.head()
```

```python
movies.drop(columns=['id'], inplace=True)
```

```python
# keep rows in ratings where movieId is in movies movieId
ratings = ratings[ratings['movieId'].isin(movies['movieId'])]
ratings.shape
```

```python
movies.head()
```

```python
movies.to_csv('data/movies_complete.csv', index=False)
```

```python
ratings.to_csv('data/ratings_complete.csv', index=False)
```
